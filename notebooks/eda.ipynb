Este notebook realiza a exploração inicial dos dados financeiros sintéticos
gerados para o treinamento dos modelos de classificação de transações e perfil financeiro.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (12, 6)
sns.set_style('whitegrid')
```

## 1. Carregamento dos Dados

```python
df_transacoes = pd.read_csv('../data/transacoes.csv')
df_perfil = pd.read_csv('../data/perfil.csv')

print('Transacoes:', df_transacoes.shape)
print('Perfil:', df_perfil.shape)
```

## 2. Análise de Transações

```python
df_transacoes.head()
```

```python
df_transacoes['categoria'].value_counts().plot(kind='bar')
plt.title('Distribuicao de Categorias')
plt.show()
```

## 3. Análise de Perfil Financeiro

```python
df_perfil['perfil_financeiro'].value_counts().plot(kind='bar')
plt.title('Distribuicao de Perfis Financeiros')
plt.show()
```

## 4. Conclusões

- O dataset de transações possui X amostras distribuídas em 8 categorias.
- O dataset de perfil possui Y amostras com distribuição entre as classes Saudavel, Em observacao e Em risco.
